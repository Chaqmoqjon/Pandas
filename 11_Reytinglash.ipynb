```python
import numpy as np
import pandas as pd
```

```python
# KEling keyingi mashqlar uchun Githubdan kattaroq DF yuklab olamiz
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/usa_cars.csv', index_col=0)
df.head()
```

```python
df['rating'] = df['year'].rank()
```

```python
df.sort_values('year')
```

```python
df['rating'] = df['price'].rank(method='min') #minimum qiymatdan boshlab reytinglash
```

```python
df['rating' ] = df['mileage'].rank(ascending=False) #bosgan yo'li bo'yicha reytinglash
```

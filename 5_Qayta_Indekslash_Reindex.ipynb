```python
import pandas as pd
```

```python
uz_data = {
    'Yil'  : [2024, 2023, 2022, 2021, 2020, 2019, 2018, 2013],
    'Aholi' : [36.1, 35.5, 34.6, 34.0, 33.5, 32.9, 32.1, 30.5],
    'Temp' : [1.54, 1.55, 1.64, 1.53, 1.68, 1.57, 1.60, 1.55],
     'Zichlik' : [76.1, 75.5, 74.6, 74.0, 73.5, 72.9, 72.1, 70.5]
}

df = pd.DataFrame(uz_data)
df
```

```python
df.index = [24, 23, 22, 21, 20, 19, 18, 13]  # index o'zgartirildi
df
```

```python
#Reindex
ind = [24, 23, 22, 21, 20, 19, 18, 13]
ind.sort()  #index tartiblandi
ind
```

```python
df.reindex(ind) # index sin tartiblab o'zgartirildi
df
```

```python
df.reindex([13, 18, 23]) # keraklilarini ajratib olish
```

```python
df.reindex(index = range(12, 25), method= 'nearest') # qiymati yo'q index lar NaN qaytaradi, "nearest" = eng yaqini bilan to'diradi, "ffill" = oxirgisi bilan to'ldiradi
```

```python
df2 = df.reindex(columns=['Yil', 'Aholi', 'Zichlik'])  # tanlangan ustunlarni ajratib beradi. .head(), sample(3)
df2.sample(3)
```

```python
df2.loc[[13, 20, 24], ['Yil', 'Aholi', 'Zichlik']]  # manzili boyicha kesib olish
```

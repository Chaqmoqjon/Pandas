```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame(np.random.randn(3, 4), columns=['Olma', 'Anor', 'Qovun', 'Anjir'], index=list('abc'))
df
```

```python
np.abs(df) #sonning modulini
```

```python
np.round(df) #yaxlitlash
```

```python
df.mean() #ustun bo'yicha o'rtacha qiymat
```

```python
df.mean(axis=1) #qator bo'yicha o'rtacha
```

```python
df.apply(np.abs) #funksiyalarni chiqarib ishlatish imkonini beradi
```

```python
df.apply(sum)
```

```python
df.apply(sum, axis=1) #qator bo'yicha
```

```python
f = lambda x: x.max() - x.min() #qisa va soddagina
```

```python
df.apply(f)
```

```python
def foo(x):
  return pd.Series([x.max(), x.min()], index = ['max', 'min'])
```

```python
df.apply(foo)
```

```python
df = pd.DataFrame(np.arange(1, 13).reshape(3, 4), columns= ['Olma', 'Anor', 'Anjir', 'Behi'], index = list('abc'))
df
```

```python
df.applymap(lambda x: x*x) # har bir elementi uchun
```

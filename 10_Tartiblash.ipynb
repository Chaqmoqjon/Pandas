```python
# KEling keyingi mashqlar uchun Githubdan kattaroq DF yuklab olamiz
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/usa_cars.csv', index_col=0)
df.head()
```

```python
df.sort_index(axis = 1, inplace = True) # qator nomlari bo'yicha tartiblash
```

```python
df
```

```python
df.sort_index(ascending = False) #teskari tartiblash
```

```python
df.sort_values(by = 'price') #aytilgan qiymatlar bo'yicha saralsh
```

```python
df.sort_values(by = ['price', 'mileage'], ascending = False) # avval price keyin mileage bo'uicha saralaydi
```

```python
df.sort_values(by = 'year') #yil bo'yicha saralash
```
